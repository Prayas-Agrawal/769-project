```python
def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum()

def query_doc_sim(query, docs_ids, encoder, wiki_embeddings, subset=10):
    query_embed = encoder.encode([query], show_progress_bar=False)
    _ids = docs_ids[:subset]
    docs_embed = [ wiki_embeddings.loc[id].values[0] for id in _ids]
    docs_embed = np.array(docs_embed)
    sims = np.matmul(docs_embed, query_embed.T)
    return sims

def doc_doc_sim(docs_ids, encoder, wiki_embeddings, subset=10):
    _ids = docs_ids[:subset]
    docs_embed = [ wiki_embeddings.loc[id].values[0] for id in _ids]
    docs_embed = np.array(docs_embed)
    sims = np.matmul(docs_embed, docs_embed.T)
    return sims 

def doc_doc_sim_query_informed(query, docs_ids, encoder, wiki_embeddings):
    query_embed = encoder.encode([query], show_progress_bar=False)
    docs_embed = [ wiki_embeddings.loc[id].values[0] for id in docs_ids]
    docs_embed = np.array(docs_embed)
    qd_probs = softmax(np.matmul(docs_embed, query_embed.T))
    qd_probs_outer = np.matmul(qd_probs, qd_probs.T)
    dd_sims = np.matmul(docs_embed, docs_embed.T)
    dd_sims = np.multiply(dd_sims, qd_probs_outer)
    return dd_sims
```

```python

from submodlib.functions.facilityLocationMutualInformation import FacilityLocationMutualInformationFunction
from submodlib.functions.graphCutMutualInformation import GraphCutMutualInformationFunction
from submodlib.functions.logDeterminantMutualInformation import LogDeterminantMutualInformationFunction

def facilityLocation_MI(n, num_queries ,ids,docs, dd_sims, qd_sims, budget=5):
    obj = FacilityLocationMutualInformationFunction(n=n, num_queries=num_queries,
                                              data_sijs=dd_sims, query_sijs=qd_sims,
                                              magnificationEta=1)
    res = obj.maximize(budget=budget, show_progress=False)
    ret_ids = [ids[i] for i,_ in res]
    ret_docs = [docs[i] for i,_ in res]
    return ret_ids, ret_docs

def logdet_MI(n, num_queries, ids, docs, dd_sims, qd_sims, budget=5):
    ex = {"magnificationEta": 1,"lambdaVal": 1e-8,"query_query_sijs": None if qd_sims is None else np.ones((1,1))}
    obj = LogDeterminantMutualInformationFunction(n=n, num_queries=num_queries,
                                              data_sijs=dd_sims, query_sijs=qd_sims, **ex)
    res = obj.maximize(budget=budget, show_progress=False)
    ret_ids = [ids[i] for i,_ in res]
    ret_docs = [docs[i] for i,_ in res]
    return ret_ids, ret_docs

def graphcut_MI(n, num_queries, ids, docs, dd_sims, qd_sims, budget=5):
    obj = GraphCutMutualInformationFunction(n=n, num_queries=num_queries,
                                              query_sijs=qd_sims)
    res = obj.maximize(budget=budget, show_progress=False)
    ret_ids = [ids[i] for i,_ in res]
    ret_docs = [docs[i] for i,_ in res]
    return ret_ids, ret_docs
    
```

```python
def calc_recall(preds, ground, k=5):
    return len(set.intersection(set(preds[:k]), set(ground)))/len(ground)

def calc_micro_ecall(preds, ground, k=2):
    return len(set.intersection(set(preds[:k]), set(ground)))/len(ground)

def calc_micro_ecall(preds, ground, k=2):
    return len(set.intersection(set(preds[:k]), set(ground)))/len(ground)

def calc_precision(preds, ground, k=2):
    return len(set.intersection(set(preds[:k]), set(ground)))/len(preds[:k])


def eval_retriever_recall(ret_model, index, num2Id, wiki_df, dataset, k=2):
    df = dataset
    recall = 0
    tot = 0
    for i, r in tqdm(df.iterrows(), total = df.shape[0]):
        preds, docs, lines = search(ret_model, index, r["claim"], num2Id, wiki_df, 5)
        ground = []
        for ann in r["evidence"]:
            for ann_data in ann:
                id = ann_data[2]
                if id != None and id not in ground:
                    if(id not in wiki_df.index):
                        continue
                    ground.append(id)
        if(len(ground) > 0):
            tot += 1
            recall += calc_recall(preds, ground)
    return recall/tot

def eval_retriever_recall_already_retrieved(dataset, k=5):
    df = dataset
    recall = 0
    tot = 0
    for i, r in tqdm(df.iterrows(), total = df.shape[0]):
        preds = r["ret_ids"]
        ground = r["evidence_ids"]
        if(len(ground) > 0):
            tot += 1
            recall += calc_recall(preds, ground, k)
    return recall/tot

def evaluate_nli_dev(nli_model, wiki_df, data):
    confusion_matrix = [[0,0,0] for i in range(3)]
    device = "cuda:0" if torch.cuda.is_available() else "cpu"
    nli_model = nli_model.to(device)
    with torch.no_grad():
        for i, r in tqdm(data.iterrows(), total=data.shape[0]):
            query = r["claim"]
            id = r["id"]
            evidence = r["evidence"]
            ids, docs = getWikiIdText(evidence, wiki_df)
            fin = query + " [SEP] " + " [SEP] ".join(docs)
            tok = tokenizer(fin, padding="max_length", truncation=True, max_length=512, return_tensors="pt").to(device)
            output = nli_model(**tok)
            preds = torch.nn.functional.softmax(output.logits, dim=1)
            labels = torch.argmax(preds, dim=1)[0].item()
            ground_label = label2Num[r["label"]]
            confusion_matrix[ground_label][labels] += 1
    fp_tp_sum = np.sum(confusion_matrix, axis=1)
    fn_tp_sum = np.sum(confusion_matrix, axis=0)
    tp_sum = confusion_matrix[0][0] + confusion_matrix[1][1] + confusion_matrix[2][2]
    micro_p = tp_sum/fp_tp_sum
    micro_r = tp_sum/fn_tp_sum
    micro_f1 =( 2*micro*p*micro_r)/(micro_p + micro_r)
    return micro_f1
        
```

```python
NUM_CONTEXT_DOCS = 5

def getWikiIdText(evidence, wiki_df):
    ids = []
    text = []
    for ann in evidence:
        for ann_data in ann:
            id = ann_data[2]
            if id != None and id not in ids:
                ids.append(id)
                _text = wiki_df.loc[id].values[0] if id in wiki_df.index else ""
                text.append(_text)
    return ids, text

class Dataset(torch.utils.data.Dataset):
    def __init__(self, data):
        self.data = data
        
    def __getitem__(self, idx):
        r = self.data.iloc[idx]
        claim = r["claim"]
#         ids = eval(r["ret_ids"])[:NUM_CONTEXT_DOCS]
#         text = eval(r["ret_docs"])[:NUM_CONTEXT_DOCS]
#         evidence = r["evidence_docs"]
        evidence = []

        hyp = " [SEP] ".join(evidence)
        fin  = claim + " [SEP] " + hyp
        tokens = tokenizer(fin, padding="max_length", truncation=True, max_length=512, return_tensors="pt")
        labels = torch.tensor(label2Num[r["label"]])
        return {'input_ids': tokens.input_ids.flatten(), 'attention_mask': tokens.attention_mask.flatten(), 'labels': labels}
    
    def __len__(self):
        return len(self.data)
        
```

```python
import re
from numpy import dot
from numpy.linalg import norm

def cos(a,b):
    return dot(a, b)/(norm(a)*norm(b))


def get_sentences(text):
    s = re.split(r"\d+\t", text)
    s = [s[i] for i in range(1, len(s))]
    return s

def get_best_sentences(model, query, ids, lines, k=5):
    embs = []
    idxs = []
    sents = []
    nums = []
    emb_query = model.encode([query], show_progress_bar=False)[0]
    for i,text in enumerate(lines):
        s = get_sentences(text)
        embs.extend([cos(model.encode([_s], show_progress_bar=False)[0], emb_query) for _s in s])
        sents.extend(s)
        idxs.extend([ids[i] for j in s])
        nums.extend([i for i in range(len(s))])
    amax = np.array(embs).argsort()[-k:][::-1]
    retSents = [sents[i] for i in amax]
    retIds = [[idxs[a], nums[a]] for a in amax]
    return retSents, retIds
        
```
